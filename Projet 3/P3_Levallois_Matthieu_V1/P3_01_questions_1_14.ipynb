### Union des tableaux de disponibilité alimentaire en Kg/personne/jour

```python
animauxDispoAlimKgProt = pd.merge(Pop, animaux_dispoAlimKgProt)
vegetauxDispoAlimKgProt = pd.merge(Pop, vegetaux_dispoAlimKgProt)

pd.concat([animauxDispoAlimKgProt, vegetauxDispoAlimKgProt])

# Création du tableau issu de l'union 
DispoAlimKgProt = pd.concat([animauxDispoAlimKgProt, vegetauxDispoAlimKgProt])

# Conversion de la disponibilité alimentaire en protéines en g/personne/jour en kg/jour

DispoAlimKgProt.rename(columns={"Population en milliers d'habitants":"A"})

DispoKgProt = DispoAlimKgProt.rename(columns={"Population en milliers d'habitants":"A"})

# Il faudrait multipier la population par 1000 car l'unité de la population est le millier d'habitants
# mais il faudrait également diviser par 1000 pour convertir les grammes en kg
# donc il n'y a pas besoin d'opérateur

DispoKgProt["Valeur"] = DispoKgProt.Valeur*DispoKgProt.A

# Suppression de la colonne "Élément"

DispoKgProt.drop(columns=["Élément"])

proj_DispoKgProt = DispoKgProt.drop(columns=["Élément"])

# Renommage des colonnes

DisKgProt = proj_DispoKgProt

DisKgProt.rename(columns={"A" : "Population en milliers d'habitants", "Valeur":"Disponibilité alimentaire en protéines en kg/jour"})

Union2 = DisKgProt.rename(columns={"A" : "Population en milliers d'habitants", "Valeur":"Disponibilité alimentaire en protéines en kg/jour"})
```

La population mondiale a été calculée question 1 également pour l'année 2013.
Pour nourrir un être humain, il faut compter 2720 kcal/jour.
Source : http://www.fao.org/nouvelle/1998/981204-f.htm

Un homme a besoin d'une quantité de protéines de 0,85 g par kg, soit 59,5 g s'il pèse 70kg.
Source : http://www.fao.org/tempref/docrep/fao/004/w0073f/w0073f02.pdf (page 17 du pdf)

Les besoins en protéines peuvent varier selon le poids, l'âge et le sexe.
Les données ne nous permettant pas de classer la population mondiale selon leur poids, leur âge et leur sexe, nous simplifierons la calcul en considérant que la population n'est composée que d'hommes de 70 kg.

```python
DispoIntMondialeVegetauxKcal = Proj_VegetauxDispoIntkcal["Disponibilité intérieure en kcal"].sum()
PopulationMondiale = Pop["Population en milliers d'habitants"].sum()*1000

# Calcul du reste de la division euclidienne de la disponibilité intérieure mondiale de végétaux en kcal
# par le produit (2720*365)

HumainsNonNourrisKcal = DispoIntMondialeVegetauxKcal%(2720*365)

# Interprétation : 
# la disponibilité intérieure mondiale de végétaux en kcal (A) est donnée pour un an
# le besoin énergétique d'un humain (B) est donné pour un jour
# Il faut donc d'abord multiplier B par 365 pour avoir le besoin énergétique pour un an
# La division de A/(B*365) donnera un nombre décimal alors que nous cherchons un nombre entier (le nombre d'humains)
# En calculant le reste de la division euclidienne de A/(B*365), nous retranchons ensuite ce résultat
# à la population totale pour obtenir le nombre d'humains nourris

#Calcul du nombre d'humains nourris

HumainsNourrisKcal = PopulationMondiale-HumainsNonNourrisKcal

print("Si toute la disponibilité intérieure mondiale de végétaux était utilisée pour de la nourriture,",
      int(HumainsNourrisKcal), "personnes seraient nourries.")
```

## Question 13 : Combien de tonnes de céréales pourraient être libérées si les USA diminuaient leur production de produits animaux de 10% ?

### Démarche:
1. Restriction du tableau "vegetaux" pour n'afficher que les aliments pour animaux des USA
2. Projection du tableau "cereales" pour n'afficher que les colonnes "Zone" et "Produit"
3. Jointure des deux tableaux résultants par les colonnes "Zone" et "Produit"
4. Calcul de 10% de la somme des valeurs de la colonne "Valeur"

```python
# Restriction de "vegetaux"

vegetauxAlimAnimaux = vegetaux[vegetaux.Élément=="Aliments pour animaux"]

vegetauxAlimAnimauxUSA = vegetauxAlimAnimaux[vegetauxAlimAnimaux.Zone== "États-Unis d'Amérique"]

# Projection de "cereales"

cerealesZoneProduit = cereales.loc[ : ,('Zone', 'Produit')]

# Jointure par colonnes "Zone" et "Produit"

AlimAnimauxUSACereales = pd.merge(vegetauxAlimAnimauxUSA,cerealesZoneProduit, on=['Zone', 'Produit'])

# Calcul de 10% de la somme des valeurs de la colonne "Valeur" de "AlimAnimauxUSACereales"

TonnesCerealesLiberees = AlimAnimauxUSACereales.Valeur.sum()*0.1

print("Si les USA diminuaient leur production de produits animaux de 10%,", TonnesCerealesLiberees, "milliers de tonnes seraient libérées.")
```

## Question 14 : En Thaïlande, quelle proportion de manioc est exportée ? Quelle est la proportion de personnes en sous-nutrition ?

```python
# Affichage des données relatives au manioc de la Thaïlande du tableau "vegetaux"

vegetauxThaïlande = vegetaux[vegetaux.Zone=="Thaïlande"]

ManiocThaïlande = vegetauxThaïlande[vegetauxThaïlande.Produit=="Manioc"]

# Restriction n'affichant que la ligne "Production"

ProdManiocThaïlande = ManiocThaïlande[ManiocThaïlande.Élément == "Production"]

# Restriction n'affichant que la ligne "Exportations - Quantité"

ExpManiocThaïlande = ManiocThaïlande[ManiocThaïlande.Élément == "Exportations - Quantité"]

# Calcul de la proportion de manioc

PropExpManiocThaïlande = (ExpManiocThaïlande.Valeur.sum()/ProdManiocThaïlande.Valeur.sum())*100

print("En Thaïlande, ",round(PropExpManiocThaïlande,2), "% de la production de manioc est exportée.")
```
