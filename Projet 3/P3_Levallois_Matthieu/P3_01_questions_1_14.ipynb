#### 	disponibilité intérieure =
- disponibilité alimentaire (en milliers de tonnes) + aliments pour animaux + semences + pertes + traitements + autres utilisations
- production + importations - exportations + variations des stocks

```python
vegetaux.head(3)
```

```python
# Création d'un tableau affichant toutes les informations liées au blé en France

Tableau2 = vegetaux[vegetaux.Zone=='France']
Tableau2 = Tableau2[Tableau2.Produit=='Blé']
Tableau2 = Tableau2.pivot(index ='Produit', columns='Élément', values='Valeur')

# Calcul de la population française

PopFrance = Pop[Pop.Zone == "France"]
PopFrance = PopFrance["Population en milliers d'habitants"].sum()*1000

# Création du tableau avec une colonne par égalité

Tableau2['disponibilité alimentaire (en milliers de tonnes) + aliments pour animaux + semences + pertes + traitements + autres utilisations'] = (Tableau2['Disponibilité alimentaire en quantité (kg/personne/an)']*PopFrance/1000000)+Tableau2['Aliments pour animaux']+Tableau2['Semences']+Tableau2['Pertes']+Tableau2['Traitement']+Tableau2['Autres utilisations (non alimentaire)']
Tableau2['production + importations - exportations + variations des stocks'] = Tableau2['Production']+Tableau2['Importations - Quantité']-Tableau2['Exportations - Quantité']+Tableau2['Variation de stock']
Tableau2= Tableau2.loc[ : ,('Disponibilité intérieure','disponibilité alimentaire (en milliers de tonnes) + aliments pour animaux + semences + pertes + traitements + autres utilisations','production + importations - exportations + variations des stocks')]

Tableau2

# Les valeurs sont égales
```

#### <span style="background-color: #FFFF00">Question 3 : Calcul (pour chaque pays et chaque produit) de la disponibilité alimentaire en kcal puis en kg par protéines</span>

```python
data = pd.concat([animaux,vegetaux,cereales])
data= pd.merge(Pop,data)

data.head(3)
```

```python
# Création du tableau 3

Tableau3 = data.pivot_table(index =['Zone','Code zone',"Population en milliers d'habitants",'Produit'],
                            columns='Élément', values='Valeur', aggfunc = sum)
Tableau3 = Tableau3.loc[ : ,('Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité de protéines en quantité (g/personne/jour)')]
Tableau3 = Tableau3.reset_index()

# Conversion puis renommage des colonnes

Tableau3['Disponibilité de protéines en quantité (g/personne/jour)'] = (Tableau3['Disponibilité de protéines en quantité (g/personne/jour)']*Tableau3["Population en milliers d'habitants"]*1000*365)/1000
Tableau3['Disponibilité alimentaire (Kcal/personne/jour)'] = Tableau3['Disponibilité alimentaire (Kcal/personne/jour)']*Tableau3["Population en milliers d'habitants"]*1000*365

Tableau3= Tableau3.rename(columns={'Disponibilité alimentaire (Kcal/personne/jour)':'Disponibilité alimentaire (Kcal/an)', 'Disponibilité de protéines en quantité (g/personne/jour)':'Disponibilité de protéines en quantité (kg/an)'})

# Suppression des valeurs incohérentes

Tableau3 = Tableau3.drop(Tableau3[Tableau3['Disponibilité alimentaire (Kcal/an)']<0].index)
Tableau3 = Tableau3.drop(Tableau3[Tableau3['Disponibilité de protéines en quantité (kg/an)']<0].index)
Tableau3 = Tableau3.drop_duplicates()

Tableau3.head(3)
```

<a id="question6"></a>
#### <span style="background-color: #FFFF00">Question 6 : Calculer la disponibilité intérieure mondiale en kcal pour les végétaux.</span>

```python
# Création du tableau 6

Tableau6 = vegetaux.pivot_table(index =['Zone','Code zone','Produit'],
                                columns='Élément', values='Valeur', aggfunc = sum)
Tableau6 = Tableau6.loc[ : ,('Disponibilité intérieure')]
Tableau6 = Tableau6.reset_index()

# Suppression des valeurs incohérentes

Tableau6 = Tableau6.drop(Tableau6[Tableau6['Disponibilité intérieure']<0].index)
Tableau6 = Tableau6.drop_duplicates()

# Jointure Tableau4 et Tableau6

Tableau6 = pd.merge(Tableau4, Tableau6, how='inner',left_on=['Zone','Produit'],right_on=['Zone','Produit'])

# Suppression de la colonne "Pourcentage de protéines"

DispIntVegeKcal = Tableau6.drop(columns=["Pourcentage de protéines"])

# Renommage des colonnes "Ratio énergie/poids en kcal/kg" et "Disponibilité intérieure en milliers de tonnes"

DispIntVegeKcal = DispIntVegeKcal.rename(columns={"Ratio énergie/poids en kcal/kg" : "A", 
                                                               "Disponibilité intérieure" : "B"})

# Le ratio étant en kcal/kg et la disponibilité intérieure en milliers de tonnes
# il faut multiplier par 1 000 000

DispIntVegeKcal["Disponibilité intérieure en kcal"] = DispIntVegeKcal.A*DispIntVegeKcal.B*1000000

# Suppression des colonnes "A" et "B"

DispIntVegeKcal = DispIntVegeKcal.drop(columns=["A", "B"])

print("La disponibilité intérieure mondiale de végétaux est de",
      DispIntVegeKcal["Disponibilité intérieure en kcal"].sum(),"kcal.")
```

<a id="question7"></a>
#### <span style="background-color: #FFFF00">Question 7 : Combien d'humains pourraient être nourris si toute la disponibilité intérieure mondiale de produits végétaux était utilisée pour de la nourriture ? Donner les résultats en terme de calories puis de protéines, et exprimer ces deux résultats en pourcentage de la population mondiale.</span>

```python
# Poids moyen d'un humain : 62 kgs
# Besoin énergétique en kcal : 2500 kcal/jour
# Besoin énergétique en quantité de protéine : 0,9 g/kg/jour

PoidsHumain = 62
BesoinHumainKcalParJour = 2500
BesoinHumainKgProtParJourParKg = 0.9


# Calcul du nombre d'humains nourris

DispoIntMondialeVegetauxKcal = DispIntVegeKcal["Disponibilité intérieure en kcal"].sum()
PopulationMondiale = Pop["Population en milliers d'habitants"].sum()*1000

HumainsNourrisKcal = DispoIntMondialeVegetauxKcal/(BesoinHumainKcalParJour*365)


# Calcul du pourcentage d'humains nourris:

PourcentHumainsNourrisKcal = (HumainsNourrisKcal/PopulationMondiale)*100


# Calcul du pourcentage de protéines de la disponibilité intérieure mondiale

Tableau6["Quantité de protéines de la disponibilité intérieure de végétaux"] = (Tableau6['Pourcentage de protéines']/100)*Tableau6['Disponibilité intérieure']*1000000

HumainsNourrisKgProt = Tableau6["Quantité de protéines de la disponibilité intérieure de végétaux"].sum()/((BesoinHumainKgProtParJourParKg/1000)*PoidsHumain*365)


# Calcul du pourcentage d'humains nourris:

PourcentHumainsNourrisKgProt = (HumainsNourrisKgProt/PopulationMondiale)*100

print("Si toute la disponibilité intérieure mondiale de végétaux était utilisée pour de la nourriture,",
      int(HumainsNourrisKcal), "personnes seraient nourries.")
print("Si toute la disponibilité intérieure mondiale de végétaux était utilisée pour de la nourriture,",
      round(PourcentHumainsNourrisKcal,2),"% de la population mondiale serait nourrie.")
print("Si toute la disponibilité intérieure mondiale de végétaux était utilisée pour de la nourriture,",
      int(HumainsNourrisKgProt), "personnes seraient nourries.")
print("Si toute la disponibilité intérieure mondiale de végétaux était utilisée pour de la nourriture,",
      round(PourcentHumainsNourrisKgProt,2),"% de la population mondiale serait nourrie.")
```

<a id="question8"></a>
#### <span style="background-color: #FFFF00">Question 8 : Combien d'humains pourraient être nourris si toute la disponibilité alimentaire de produits végétaux, la nourriture végétale destinée aux animaux et les pertes de produits végétaux étaient utilisées pour de la nourriture ? Donner les résultats en terme de calories puis de protéines, et exprimer ces deux résultats en pourcentage de la population mondiale.</span>

```python
# Création du tableau 8

Tableau8 = vegetaux.pivot_table(index =['Zone','Code zone','Produit'],columns='Élément', values='Valeur', aggfunc = sum)
Tableau8 = Tableau8.loc[ : ,('Disponibilité alimentaire en quantité (kg/personne/an)','Aliments pour animaux','Pertes')]
Tableau8 = Tableau8.reset_index()

# Suppression des valeurs incohérentes

Tableau8 = Tableau8.drop(Tableau8[Tableau8['Disponibilité alimentaire en quantité (kg/personne/an)']<0].index)
Tableau8 = Tableau8.drop(Tableau8[Tableau8['Aliments pour animaux']<0].index)
Tableau8 = Tableau8.drop(Tableau8[Tableau8['Pertes']<0].index)
Tableau8 = Tableau8.drop_duplicates()
Tableau8 = Tableau8.fillna(0)

# Jointure avec Pop

Tableau8 = pd.merge(Pop,Tableau8)
Tableau8['Disponibilité alimentaire en quantité (kg/personne/an)'] = Tableau8['Disponibilité alimentaire en quantité (kg/personne/an)']*Tableau8["Population en milliers d'habitants"]*1000/1000000

#Calcul Disponibilité totale en milliers de tonnes

Tableau8['Disponibilité alimentaire en quantité+semences+pertes'] = Tableau8['Disponibilité alimentaire en quantité (kg/personne/an)']+Tableau8['Aliments pour animaux']+Tableau8['Pertes']

# Jointure avec Pop

Tableau8 = pd.merge(Tableau8,Tableau4)

# Conversion en kcal

Tableau8['Disponibilité alimentaire en quantité+semences+pertes en kcal'] = Tableau8['Disponibilité alimentaire en quantité+semences+pertes']*Tableau8['Ratio énergie/poids en kcal/kg']*1000000

# Conversion en kg de protéines

Tableau8['Disponibilité alimentaire en quantité+semences+pertes en kg de protéines'] = Tableau8['Disponibilité alimentaire en quantité+semences+pertes']*(Tableau8['Pourcentage de protéines']/100)*1000000

DispoSemencesPertesKcal = Tableau8['Disponibilité alimentaire en quantité+semences+pertes en kcal'].sum()
DispoSemencesPertesKgProt = Tableau8['Disponibilité alimentaire en quantité+semences+pertes en kg de protéines'].sum()

HumainsNourrisKcal_8 = DispoSemencesPertesKcal/(BesoinHumainKcalParJour*365)
PourcentHumainsNourrisKcal_8 = (HumainsNourrisKcal_8/PopulationMondiale)*100
HumainsNourrisKgProt_8 = DispoSemencesPertesKgProt/((BesoinHumainKgProtParJourParKg/1000)*PoidsHumain*365)
PourcentHumainsNourrisKgProt_8 = (HumainsNourrisKgProt_8/PopulationMondiale)*100

print("Si toute la disponibilité alimentaire en quantité, les semences et les pertes mondiale de végétaux étaient utilisées pour de la nourriture :")
print("En terme de besoin en kcal :")
print("-",int(HumainsNourrisKcal_8), "d'êtres humains seraient nourris.")
print("- soit",round(PourcentHumainsNourrisKcal_8,2),"% de la population mondiale.")
print("En terme de besoin en protéines :")
print("-", int(HumainsNourrisKgProt_8), "d'êtres humains seraient nourris.")
print("- soit",round(PourcentHumainsNourrisKgProt_8,2),"% de la population mondiale serait nourrie.")
```

<a id="réponse12_5"></a>
#### Tableau pour chaque produit des ratios : 
- entre quantité destinés aux "Autres utilisations" et la disponibilité intérieure
- entre quantité destinés à la nourriture animale et celle de la nourriture totale (animale + humaine)

```python
# Restriction de "data" n'affichant que les "Autres utilisations (non alimentaire)"

AutresUtilisationsMondiale = data[data.Élément == "Autres utilisations (non alimentaire)"]

# Restriction n'affichant que la disponibilité intérieure des 15 produits les plus exportés
# par les pays recensant des populations sous-alimentées

AutresUtilisationsMondiale15Produits = pd.merge(ProduitsTop15Exportations, AutresUtilisationsMondiale)
AutresUtilisationsMondiale15Produits = AutresUtilisationsMondiale15Produits.drop_duplicates()
AutresUtilisationsMondiale15Produits = AutresUtilisationsMondiale15Produits.loc[ : , ('Produit', 'Zone', 'Valeur')]
AutresUtilisationsMondiale15Produits = AutresUtilisationsMondiale15Produits.groupby(['Produit']).sum()

# Renommage de la colonne "Valeur" pour opération

AutresUtilisationsMondiale15Produits = AutresUtilisationsMondiale15Produits.rename(columns={'Valeur':'A'})

# Restriction de "data" n'affichant que la disponibilité intérieure

DispIntMondiale = data[data.Élément == "Disponibilité intérieure"]

# Restriction n'affichant que la disponibilité intérieure des 15 produits les plus exportés
# par les pays recensant des populations sous-alimentées

DispIntMondiale15Produits = pd.merge(ProduitsTop15Exportations, DispIntMondiale)
DispIntMondiale15Produits = DispIntMondiale15Produits.drop_duplicates()
DispIntMondiale15Produits = DispIntMondiale15Produits.drop(DispIntMondiale15Produits[DispIntMondiale15Produits.Valeur<0].index)
DispIntMondiale15Produits = DispIntMondiale15Produits.loc[ : , ('Produit', 'Zone', 'Valeur')]
DispIntMondiale15Produits = DispIntMondiale15Produits.groupby(['Produit']).sum()

# Renommage de la colonne "Valeur" pour opération

DispIntMondiale15Produits = DispIntMondiale15Produits.rename(columns={'Valeur':'B'})

# Restriction de "data" n'affichant que les "Aliments pour animaux"

AlimentsAnimauxMondiale = data[data.Élément == "Aliments pour animaux"]

# Restriction n'affichant que la disponibilité intérieure des 15 produits les plus exportés
# par les pays recensant des populations sous-alimentées

AlimentsAnimauxMondiale15Produits = pd.merge(ProduitsTop15Exportations, AlimentsAnimauxMondiale)
AlimentsAnimauxMondiale15Produits = AlimentsAnimauxMondiale15Produits.drop_duplicates()
AlimentsAnimauxMondiale15Produits = AlimentsAnimauxMondiale15Produits.loc[ : , ('Produit', 'Zone', 'Valeur')]
AlimentsAnimauxMondiale15Produits = AlimentsAnimauxMondiale15Produits.groupby(['Produit']).sum()

# Renommage de la colonne 'Valeur' pour opération

AlimentsAnimauxMondiale15Produits = AlimentsAnimauxMondiale15Produits.rename(columns={'Valeur':'C'})

# Restriction de "data" n'affichant que les éléments "Nourriture"

NourritureMondiale = data[data.Élément == "Nourriture"]

# Restriction n'affichant que la disponibilité intérieure des 15 produits les plus exportés
# par les pays recensant des populations sous-alimentées

NourritureMondiale15Produits = pd.merge(ProduitsTop15Exportations, NourritureMondiale)
NourritureMondiale15Produits = NourritureMondiale15Produits.drop_duplicates()
NourritureMondiale15Produits = NourritureMondiale15Produits.loc[ : , ('Produit', 'Zone', 'Valeur')]
NourritureMondiale15Produits = NourritureMondiale15Produits.groupby(['Produit']).sum()

# Renommage de la colonne "Valeur" pour opération

NourritureMondiale15Produits = NourritureMondiale15Produits.rename(columns={'Valeur':'D'})

# Jointure de "AutresUtilisationsMondiale15Produits" et"DispIntMondiale15Produits"

Jointure1 = pd.merge(AutresUtilisationsMondiale15Produits, DispIntMondiale15Produits, on='Produit')
Jointure2 = pd.merge(AlimentsAnimauxMondiale15Produits, NourritureMondiale15Produits, on='Produit')

Tableau12 = pd.merge(Jointure1,Jointure2, on='Produit')
Tableau12['Ratio  "Autres utilisations" / Disponibilité intérieure'] = Tableau12.A/Tableau12.B
Tableau12['Ratio  Aliments pour animaux / Nourriture totale (animale + humaine)'] = Tableau12.C/(Tableau12.C+Tableau12.D)
Tableau12 = Tableau12.drop(columns=['A','B','C','D'])

Tableau12.head(3)
```
